## Train

```python
import os
import sys
sys.path.append("./monk_v1/monk/");
import psutil
from pytorch_prototype import prototype
```

```python
ptf = prototype(verbose=1);
ptf.Prototype("indoor_scene", "exp1");
```

```python
ptf.Default(dataset_path="./Train/", model_name="resnet101", freeze_base_network=True, num_epochs=10);
```

```python
ptf.Train()
```

## Inference

```python
import numpy as np
import PIL.Image as Image
import requests

test_url = "https://i.ytimg.com/vi/RfYc0BUqkMs/maxresdefault.jpg"

# URL of the image to be downloaded is defined as image_url 
r = requests.get(test_url) # create HTTP response object 
with open('test.jpg','wb') as f: 
    f.write(r.content) 
test = Image.open('test.jpg')
test
```

```python
ptf.Prototype("indoor_scene", "exp1", eval_infer=True);
```

```python
img_name = "test.jpg";
predictions = ptf.Infer(img_name=img_name, return_raw=False);
print(predictions)
```

## Fine-Tuning

```python
ptf = prototype(verbose=1);
ptf.Prototype("indoor_scene", "exp2", copy_from=["indoor_scene", "exp1"]);
```

```python
########Update Transforms#####################
ptf.apply_random_resized_crop(224, train=True, val=True);
###############################################
```

```python
##########Update hyperparameters################
ptf.optimizer_asgd(0.001, weight_decay=0.00001);
ptf.update_num_epochs(10);
################################################
```

```python
###########For freeze layers#####################
ptf.update_freeze_layers(100); 
################################################
```

```python
ptf.Reload()
```

```python
ptf.Train()
```

## Resume Training

```python
ptf = prototype(verbose=1)
ptf.Prototype("indoor_scene", "exp2", resume_train=True);
```

```python
ptf.Train()
```
